```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Dataset
file_path = '/content/sample_data/california_housing_train.csv'
df = pd.read_csv(file_path)

# 2. Line Plot (Housing Median Age vs Median House Value)
plt.figure(figsize=(10, 5))
plt.plot(df['housing_median_age'][:50], color='blue', label='Median Age')
plt.title('Housing Median Age (First 50 samples)')
plt.xlabel('Index')
plt.ylabel('Age')
plt.legend()
plt.show()

# 3. Scatter Plot (Longitude vs Latitude)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='longitude', y='latitude', data=df, hue='median_house_value')
plt.title('Geographical Distribution of House Values')
plt.show()

# 4. Histogram (Distribution of Population)
plt.figure(figsize=(8, 5))
sns.histplot(df['population'], bins=30, color='green', kde=True)
plt.title('Population Distribution')
plt.xlabel('Population')
plt.show()
```
