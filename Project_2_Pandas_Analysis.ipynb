```python
import pandas as pd

# 1. Loading Dataset
file_path = '/content/sample_data/california_housing_train.csv'
df = pd.read_csv(file_path)

# 2. Data Inspection
print("--- Dataset Head ---")
print(df.head())

print("\n--- Dataset Info ---")
print(df.info())

# 3. Statistical Summary
print("\n--- Statistics ---")
print(df.describe())

# 4. Selection and Filtering
selected_data = df[['total_rooms', 'population', 'median_house_value']]
print("\n--- Selected Columns ---")
print(selected_data.head())

filtered_data = df[df['total_rooms'] > 1500]
print(f"\nRows with rooms > 1500: {len(filtered_data)}")
```
